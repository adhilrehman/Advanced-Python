```python
invalid_count = 0
valid_count = 0
valid_nums = set()

numbers=['9876543210','12453','9988776655','9876501234']
for num in numbers:
  if len(num) != 10:
    print('invalid')
    invalid_count += 1
  else:
    print('valid')
    valid_count += 1
    valid_nums.add(num)

print(f"invalid count is {invalid_count} and valid is {valid_count}")
print(f"valid_nums: {valid_nums}")
```
